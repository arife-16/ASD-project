```python
import os, subprocess
DRIVE_ROOT = '/content/drive/MyDrive/ASD'
PHENO_CSV = os.path.join(DRIVE_ROOT, 'phenotype.csv')
NIFTI_DIR = os.path.join(DRIVE_ROOT, 'nifti')
ATLAS_PATH = os.path.join(DRIVE_ROOT, 'atlas', 'atlas.nii.gz')
TSV_DIR = os.path.join(DRIVE_ROOT, 'confounds_tsv')
WM_MASK = os.path.join(DRIVE_ROOT, 'masks', 'wm.nii.gz')
CSF_MASK = os.path.join(DRIVE_ROOT, 'masks', 'csf.nii.gz')
OUT_DIR = os.path.join(DRIVE_ROOT, 'experiment_results')
os.makedirs(OUT_DIR, exist_ok=True)
strategies = ['stratified', 'group', 'site_stratified', 'loso']
for cv in strategies:
    cmd = [
        'python', '/content/ASD-project/scripts/run_pipeline.py',
        '--phenotype', PHENO_CSV,
        '--nifti_dir', NIFTI_DIR,
        '--atlas', ATLAS_PATH,
        '--atlas_type', 'labels',
        '--site_col', 'SITE',
        '--label_col', 'DX_GROUP',
        '--age_col', 'AGE_AT_SCAN',
        '--sex_col', 'SEX',
        '--tr', '2.0',
        '--window_size', '50',
        '--step', '10',
        '--n_states', '5',
        '--cv_strategy', cv,
        '--tune_models',
        '--confounds_tsv_dir', TSV_DIR,
        '--wm_mask', WM_MASK,
        '--csf_mask', CSF_MASK,
        '--output', os.path.join(OUT_DIR, f'results_{cv}.json')
    ]
    subprocess.run(cmd, check=False)
```
